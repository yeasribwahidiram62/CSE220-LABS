```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
def sum_of_leaves(root, sum):
    if root == None:
        return 0
    else:
        iram_lt = sum_of_leaves(root.left, sum)
        iram_rt = sum_of_leaves(root.right, sum)
        if iram_lt == 0 and iram_rt == 0:
            return root.elem
        return iram_lt + iram_rt



#DRIVER CODE
#Write by yourself from the given tree
root = BTNode(30)
root.left = BTNode(10)
root.left.left = BTNode(3)
root.left.left.left = BTNode(2)
root.left.right = BTNode(15)
root.right = BTNode(40)
root.right.left = BTNode(35)
root.right.left.right = BTNode(36)
root.right.right = BTNode(55)

print(sum_of_leaves(root, 0))
```
